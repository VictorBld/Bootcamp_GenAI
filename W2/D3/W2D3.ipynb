```python
import pandas as pd

train=pd.read_csv('train.csv', header=0)

train.head()
```

```python
print(train.duplicated().sum())
```

```python
train=train.drop_duplicates()

print(train.duplicated().sum())
```

```python
train.info()
```

```python
missing_data=train.isnull().sum()
missing_data
```

Theres is missing values in : "Age","Cabin" and "Embarked"

We have different ways to handle missing values. For the Cabin column, we can drop it because the information is not important enough for our model.

For the Age column, I decided to fill the missing values with the mean of the entire column to retain all the data.

Finally, I chose to drop the two rows with missing values in Embarked because if we don't know where these passengers boarded, we can do without that information. Moreover, removing just two rows won't have a significant impact on our model.

```python
train_filled=train
train_filled["Age"]=train_filled["Age"].fillna(train_filled["Age"].mean())
train_filled.info()
```

```python
train_filled.isnull().sum()
```

```python
train_filled.dropna(subset=["Embarked"], inplace=True)
train_filled.info()
```

```python
train_filled.drop("Cabin", axis=1, inplace=True)
train_filled.info()
```

```python
train_filled["FamilySize"]=train_filled["SibSp"]+train_filled["Parch"]+1
train_filled.info()
```

```python
train_filled["Name"]
train_filled["Title"] = train_filled["Name"].apply(lambda Name: Name.split(",")[1].split(".")[0].strip())
train_filled["Title"].value_counts()
```

```python
title_mapping = {
    "Mr": "Mr", "Mrs": "Mrs", "Miss": "Miss", "Master": "Master",
    "Dr": "Rare", "Rev": "Rare", "Col": "Rare", "Major": "Rare",
    "Mlle": "Miss", "Ms": "Miss", "Mme": "Mrs", "Countess": "Rare",
    "Lady": "Rare", "Sir": "Rare", "Jonkheer": "Rare", "Don": "Rare", "Capt": "Rare"
}
train_filled["Title"] = train_filled["Title"].map(title_mapping)
train_filled["Title"].value_counts()
```

```python
train_filled["Sex"]=train_filled["Sex"].map({"male" : 0, "female" : 1})
train_filled["Sex"].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
train_filled["Title"] = LabelEncoder().fit_transform(train_filled["Title"])
```

```python
train_filled['Title'].value_counts()
```

```python
train_filled["Embarked"].value_counts()
train_filled["Embarked"]=train_filled["Embarked"].map({"S" : 0, "C" : 1, "Q" : 2})
train_filled["Embarked"].value_counts()
```

```python
train_filled.info()
train_filled.head()
```

```python
#Detection d'outliers
Q1= train_filled["Age"].quantile(0.25)
Q3= train_filled["Age"].quantile(0.75)
IQR= Q3-Q1
lower_bound= Q1-1.5*IQR
upper_bound= Q3+1.5*IQR
outlier=train[((train_filled["Age"] < lower_bound) | (train_filled["Age"] > upper_bound))]
```

```python
print(f"le nombre d'oulier est : {outlier.shape[0]}")
```

```python
train_filled=train_filled[((train_filled["Age"] >= lower_bound) & (train_filled["Age"] <= upper_bound))]
```

```python
train_filled.info()
```

```python
numerical_cols = ['Age', 'Fare', 'SibSp', 'Parch', 'FamilySize']
for col in numerical_cols:
    print(f"Column: {col}")
    print(f"  Min: {train_filled[col].min()}")
    print(f"  Max: {train_filled[col].max()}")
    print(f"  Mean: {train_filled[col].mean()}")
    print(f"  Standard Deviation: {train_filled[col].std()}")
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
scaler = StandardScaler()
train_filled[['Age', 'Fare']] = scaler.fit_transform(train_filled[['Age', 'Fare']])
```

```python
min_max_scaler = MinMaxScaler()
train_filled[['SibSp', 'Parch', 'FamilySize']] = min_max_scaler.fit_transform(train_filled[['SibSp', 'Parch', 'FamilySize']])
```

```python
train_filled.head()
```

```python
train_filled['AgeGroup'] = pd.cut(train_filled['Age'], bins=[0, 18, 35, 60, 100], labels=['Child', 'YoungAdult', 'Adult', 'Senior'])
```

```python
age_dummies = pd.get_dummies(train_filled['AgeGroup'], prefix='AgeGroup')
train_filled = pd.concat([train_filled, age_dummies], axis=1)
```

```python
train_filled.drop('AgeGroup', axis=1, inplace=True)
```
