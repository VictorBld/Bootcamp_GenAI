Data visualization makes complex data easier to understand, reveals trends and patterns, enhances decision-making, communicates insights clearly, and keeps audiences engaged.

A line graph shows trends over time, compares multiple variables, highlights patterns, reveals relationships, and helps forecast future values.

```python
import matplotlib.pyplot as plt
import pandas as pd

temperatures = [72, 74, 76, 80, 82, 78, 75]
days = [1, 2, 3, 4, 5, 6, 7]

plt.plot(days, temperatures)
plt.xlabel('Day')
plt.ylabel('Temperature (F)')
plt.title('Temperature Over Time')
plt.show()
```

```python
sales =[5000, 5500, 6200, 7000, 7500]
months = [1,2,3,4,5]

plt.bar(months, sales)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.show()
```

```python
import seaborn as sns


data = {'CGPA': ['2.0-2.5', '2.5-3.0', '3.0-3.5', '3.5-4.0'],
        'Number of Students': [10, 25, 30, 15]}
df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='CGPA', weights='Number of Students', color='skyblue')
plt.title('Distribution of Student CGPA')
plt.xlabel('CGPA Range')
plt.ylabel('Number of Students')
plt.show()
```

```python
data = {'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
        'Anxiety Level': ['Non', 'Non', 'Oui', 'Non', 'Oui']}

df= pd.DataFrame(data)

df
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Anxiety Level', hue='Gender', data=df, palette='viridis')
plt.title('Anxiety Levels Across Genders')
plt.xlabel('Anxiety Level')
plt.ylabel('Count')
plt.show()
```

```python
data = {'Age': [18, 19, 20, 21, 22, 18, 19, 20, 21, 22],
        'Do you have Panic Attacks?': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes']}
df = pd.DataFrame(data)

df['Panic Attacks'] = df['Do you have Panic Attacks?'].map({'Yes': 1, 'No': 0})

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Panic Attacks', data=df, hue='Panic Attacks', palette=['red', 'blue'], s=100)

plt.title('Relationship between Age and Panic Attacks')
plt.xlabel('Age')
plt.ylabel('Panic Attacks (1=Yes, 0=No)')
plt.xticks(range(min(df['Age']), max(df['Age']) + 1))
plt.yticks([0, 1])

plt.show()
```
