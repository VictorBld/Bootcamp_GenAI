```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

df= pd.read_csv('2019.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
print(df.isnull().sum())
```

```python
plt.scatter(df['Social support'], df['Score'])
plt.xlabel('Social support')
plt.ylabel('Score')
plt.title('Relationship between Social support and Score')
plt.grid(True)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Subplot 1: GDP per Capita by region
sb.scatterplot(ax=axes[0], x='Country or region', y='GDP per capita', data=df)
axes[0].set_xlabel('Country or region')
axes[0].set_ylabel('GDP per capita')
axes[0].set_title('GDP per Capita by Region')
axes[0].grid(True)

# Subplot 2: Distribution of Healthy Life Expectancy by Region
sb.boxplot(ax=axes[1], x='Country or region', y='Healthy life expectancy', data=df)
axes[1].set_xlabel('Country or region')
axes[1].set_ylabel('Healthy Life Expectancy')
axes[1].set_title('Healthy Life Expectancy Distribution by Region')
axes[1].grid(True)
axes[1].tick_params(axis='x', rotation=45)


plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.bar(df['Country or region'], df['GDP per capita'], color='skyblue', label='GDP per Capita')
ax1.set_xlabel('Country or Region')
ax1.set_ylabel('GDP per Capita', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax1.tick_params(axis='x', rotation=90)

ax2 = ax1.twinx()
ax2.plot(df['Country or region'], df['Healthy life expectancy'], color='green', label='Healthy Life Expectancy')
ax2.set_ylabel('Healthy Life Expectancy', color='coral')
ax2.tick_params(axis='y', labelcolor='coral')

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

plt.title('GDP per Capita vs. Healthy Life Expectancy')
plt.tight_layout()
plt.show()
```
