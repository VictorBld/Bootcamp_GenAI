```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

df=pd.read_excel("US Superstore data.xls")
```

```python
df.head()
```

```python
print(df.isnull( ).sum())
```

There is no missing values in our dataset.

```python
df['Order Date']= pd.to_datetime(df['Order Date'])

df['Year'] = df ['Order Date'].dt.year

yearly_sales= df.groupby('Year')['Sales'].sum()

plt.figure(figsize=(10,6))
sns.lineplot(x=yearly_sales.index, y=yearly_sales.values)

plt.title('Sales trend over years')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.grid(True)

plt.show()
```

```python
df=pd.read_excel("US Superstore data.xls")

product_sales = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=product_sales.values, y=product_sales.index)
plt.title('Top 10 Products by Sales')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title('Relationship between Profit and Discount')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```

I feel like Matplotlib offer more basic informations through flexible graph. I'd use it more for precise, detailed and mathematical visualizations

In the other hand Seaborn gives more sophisticated  statistical insights with a simpler syntax and a beautiful aesthetics. Better to explore relationships and distributions within data.
