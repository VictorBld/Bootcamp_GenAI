```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

temperature_data = np.random.uniform(low=-5, high=35, size=(10,12))

temperature_data
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
cities = ['City A', 'City B', 'City C', 'City D', 'City E', 'City F', 'City G', 'City H', 'City I', 'City J']
temperature_df = pd.DataFrame(temperature_data, index=cities, columns=months)

temperature_df
```

```python
annual_average_temperatures = temperature_df.mean(axis=1)

city_with_highest_avg_temp = annual_average_temperatures.idxmax()

city_with_lowest_avg_temp = annual_average_temperatures.idxmin()

print("Annual average temperatures:\n", annual_average_temperatures)
print("\nCity with highest average temperature:", city_with_highest_avg_temp)
print("City with lowest average temperature:", city_with_lowest_avg_temp)
```

```python
plt.figure(figsize=(12, 6))
for city in cities:
    plt.plot(months, temperature_df.loc[city], label=city)

plt.xlabel("Month")
plt.ylabel("Temperature (Â°C)")
plt.title("Monthly Temperature Evolution for Cities")
plt.legend()
plt.grid(True)
plt.show()
```
