```python
import numpy as np

# Example survey data matrix (5 respondents, 4 questions)
survey_data = np.array([[5, 1, 4, 4],
                        [4, 2, 4, 3],
                        [5, 1, 1, 5],
                        [3, 2, 1, 2],
                        [2, 1, 2, 5]])

# Calculating various statistics
average_responses = np.mean(survey_data, axis=0)
total_responses = np.sum(survey_data, axis=0)
max_responses = np.max(survey_data, axis=0)
min_responses = np.min(survey_data, axis=0)

print("Average Responses:", average_responses)
print("Total Responses:", total_responses)
print("Maximum Responses:", max_responses)
print("Minimum Responses:", min_responses)
```

```python
import matplotlib.pyplot as plt

# Generate a normal distribution
mean = 0
std_dev = 1
samples = np.random.normal(mean, std_dev, 1000)

# Plotting the distribution
plt.hist(samples, bins=30, density=True)
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Random sampling from an array
data = np.arange(1,101)
sample = np.random.choice(data, size=5, replace=False)
print("Random Sample:", sample)
print(data)
```

```python
# Shuffling data
np.random.shuffle(data)
print("Shuffled Data:", data)
```

```python
mean=np.mean(data)
print("Mean : ",mean)
median=np.median(data)
print("Median : ",median)
variance=np.var(data)
print("Variance : ",variance)

from scipy.stats import skew, kurtosis

print("skeewness:", skew(data))
print("kurtosis:", kurtosis(data))
```

```python
import pandas as pd

date_range = pd.date_range(start='2022-01-01', end='2022-12-31', freq='D')
print(date_range)
```

```python
temperature_data = np.random.uniform(10, 35, size=len(date_range))
print(temperature_data)
```

```python
humidity_data = np.random.uniform(30, 90, size=len(date_range))
print(humidity_data)
```

```python
weather_df = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature_data,
    'Humidity': humidity_data
})

print(weather_df)
```

```python
mean_temp = weather_df['Temperature'].mean()
mean_humidity = weather_df['Humidity'].mean()

median_temp = weather_df['Temperature'].median()
median_humidity = weather_df['Humidity'].median()

std_temp = weather_df['Temperature'].std()
std_humidity = weather_df['Humidity'].std()


print("Mean Temperature:", mean_temp)
print("Mean Humidity: ",mean_humidity)
print("Median Temperature:" ,median_temp)
print("Median Humidity: ",median_humidity)
print("Standard Deviation Temperature: ",std_temp)
print("Standard Deviation Humidity:",std_humidity)
```

```python
weather_df.describe()
```

```python
# Date with the highest temperature
max_temp_date = weather_df.loc[weather_df['Temperature'].idxmax(), 'Date']
print("Date with the highest temperature:", max_temp_date)

# Date with the highest humidity
max_humidity_date = weather_df.loc[weather_df['Humidity'].idxmax(), 'Date']
print("Date with the highest humidity:", max_humidity_date)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))
plt.plot(weather_df['Date'], weather_df['Temperature'], label='Temperature', color='r')
plt.plot(weather_df['Date'], weather_df['Humidity'], label='Humidity', color='b')
plt.title('Temperature and Humidity Trend Over the Year')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```
