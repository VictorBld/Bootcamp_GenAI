```python
numerical_stats = df_clean.describe()
print(numerical_stats)

country_counts = df_clean['country'].value_counts()
print("\nDistribution of power plants by country:\n", country_counts)

fuel_counts = df_clean['primary_fuel'].value_counts()
print("\nDistribution of power plants by fuel type:\n", fuel_counts)

country_fuel_counts = df_clean.groupby(['country', 'primary_fuel']).size().unstack(fill_value=0)
print("\nDistribution of power plants by country and fuel type:\n", country_fuel_counts)
```

```python
from scipy import stats

# Group data by fuel type
fuel_groups = df_clean.groupby('primary_fuel')['capacity_mw']

# Calculate mean and standard deviation for each fuel type
means = fuel_groups.mean()
std_devs = fuel_groups.std()

print("\nMean power output by fuel type:\n", means)
print("\nStandard deviation of power output by fuel type:\n", std_devs)


fuel_types_to_compare = ['Gas', 'Oil', 'Coal']
data_for_anova = [df_clean['capacity_mw'][df_clean['primary_fuel'] == fuel] for fuel in fuel_types_to_compare]

f_statistic, p_value = stats.f_oneway(*data_for_anova)

print("\nF-statistic:", f_statistic)
print("P-value:", p_value)

alpha = 0.05  # Significance level

if p_value < alpha:
    print("\nThere is a statistically significant difference in mean power output among the selected fuel types.")
else:
    print("\nThere is no statistically significant difference in mean power output among the selected fuel types.")


# Perform t-tests for pairwise comparisons (example: 'Gas' vs 'Oil')

fuel1 = 'Gas'
fuel2 = 'Oil'

group1 = df_clean['capacity_mw'][df_clean['primary_fuel'] == fuel1]
group2 = df_clean['capacity_mw'][df_clean['primary_fuel'] == fuel2]

t_statistic, p_value_t = stats.ttest_ind(group1, group2)

print(f"\nT-test results for {fuel1} vs {fuel2}:")
print("T-statistic:", t_statistic)
print("P-value:", p_value_t)

if p_value_t < alpha:
  print(f"\nThere is a statistically significant difference in mean power output between {fuel1} and {fuel2}.")
else:
  print(f"\nThere is no statistically significant difference in mean power output between {fuel1} and {fuel2}.")
```

```python
import numpy as np
# Convert commissioning year to numeric, coercing errors to NaN
df_clean['commissioning_year'] = pd.to_numeric(df_clean['commissioning_year'], errors='coerce')
# Analyze trends in fuel type mix over time
fuel_trends = df_clean.groupby(['commissioning_year', 'primary_fuel']).size().unstack(fill_value=0)

# Example: Calculate the proportion of each fuel type over time
fuel_proportions = fuel_trends.div(fuel_trends.sum(axis=1), axis=0)
print("\nProportions of each fuel type over commissioning years:\n", fuel_proportions)

#Further analysis using numpy for specific trends. e.g.
#For example, you could calculate the moving average of the proportions.

    # Example: Calculate the 5-year moving average for 'Gas'
if 'Gas' in fuel_proportions.columns:
  gas_proportions = fuel_proportions['Gas'].fillna(0)
  window_size = 5
  gas_moving_average = np.convolve(gas_proportions, np.ones(window_size), 'valid') / window_size
  print("\n5-year moving average of Gas proportion:\n", gas_moving_average)


else:
    print("No 'commissioning_year' column found in the dataset.")
```

```python
matrix_data = df_clean[['capacity_mw', 'latitude', 'longitude']].dropna()
# Convert the selected columns to a NumPy matrix
power_matrix = matrix_data.values

# Calculate the covariance matrix
covariance_matrix = np.cov(power_matrix, rowvar=False)

# Calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

print("Eigenvalues:\n", eigenvalues)
print("\nEigenvectors:\n", eigenvectors)

# Analyze the results. For example:
# 1. Eigenvalues represent the variance along each principal component.
#    Larger eigenvalues indicate principal components that capture more variance in the data.

# 2. Eigenvectors represent the directions of these principal components.

#In the context of power plant attributes, these principal components would capture linear combinations of capacity, latitude, and longitude.
#The eigenvectors  provide insights into the relationships between these attributes.
#A high eigenvalue associated with an eigenvector pointing primarily along the capacity axis would indicate that capacity varies more than the other attributes.
#An eigenvector that has nearly equal weights for latitude, longitude, and capacity could indicate some correlation between geographic location and capacity.


# Example: Identify the principal component with the largest eigenvalue
max_eigenvalue_index = np.argmax(eigenvalues)
dominant_eigenvector = eigenvectors[:, max_eigenvalue_index]

print("\nDominant Eigenvector (corresponding to the largest eigenvalue):\n", dominant_eigenvector)

#Interpretation:
#The dominant eigenvector shows the relative importance of each attribute (capacity, latitude, longitude) in the direction of the greatest variance.
#Examine the magnitude of elements in the eigenvector to understand the relative influence of each attribute.

```
