DATA CLEANING AND PREPROCESSING

No missing data & only float/integer so we don't need to do anything

```python
# Calculate central tendency measures for each numerical feature
for col in df.select_dtypes(include=['number']):
    print(f"Feature: {col}")
    print(f"  Mean: {df[col].mean()}")
    print(f"  Median: {df[col].median()}")
    print(f"  Mode: {df[col].mode()[0]}\n") # Access the first mode value

# Dispersion measures (range, variance, standard deviation, IQR)
for col in df.select_dtypes(include=['number']):
    print(f"Feature: {col}")
    print(f"  Range: {df[col].max() - df[col].min()}")
    print(f"  Variance: {df[col].var()}")
    print(f"  Standard Deviation: {df[col].std()}")
    print(f"  IQR: {df[col].quantile(0.75) - df[col].quantile(0.25)}\n")


# Shape (Skewness and Kurtosis)
for col in df.select_dtypes(include=['number']):
    print(f"Feature: {col}")
    print(f"  Skewness: {df[col].skew()}")
    print(f"  Kurtosis: {df[col].kurt()}\n")

```

```python
corr= df.corr()
corr
# sns.heatmap(corr, annot=True,linewidths=0.9)
# plt.show()
```

```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr


# Calculate Pearson correlation between numerical features and target variable ('price_range')
for col in df.select_dtypes(include=['number']):
  if col != 'price_range':  # Exclude the target variable itself
    correlation, p_value = pearsonr(df[col], df['price_range'])
    print(f"Pearson Correlation between {col} and price_range: {correlation:.3f}, P-value: {p_value:.3f}")

print("\n")

# Calculate Spearman rank correlation
for col in df.select_dtypes(include=['number']):
  if col != 'SalePrice':
    correlation, p_value = spearmanr(df[col], df['price_range'])
    print(f"Spearman Rank Correlation between {col} and price_range: {correlation:.3f}, P-value: {p_value:.3f}")
```

```python
from scipy.stats import kendalltau

for col in df.select_dtypes(include=['number']):
  if col != 'price_range':
    correlation, p_value = kendalltau(df[col], df['price_range'])
    print(f"Kendall's Tau Correlation between {col} and price_range: {correlation:.3f}, P-value: {p_value:.3f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



plt.figure(figsize=(8, 6))
sns.histplot(df['price_range'], kde=True)
plt.title(f'Distribution of price_range')
plt.xlabel('price_range')
plt.ylabel('Frequency')
plt.show()

# Scatter plots (example: 'price_range' vs. 'battery_power')
plt.figure(figsize=(8, 6))
sns.scatterplot(x='battery_power', y='price_range', data=df)
plt.title('Relationship between price_range and battery_power')
plt.xlabel('barrery_power')
plt.ylabel('price_range')
plt.show()

# Box plots (example: 'price_range' across different 'n_cores' categories)
plt.figure(figsize=(10, 6))
sns.boxplot(x='n_cores', y='price_range', data=df)
plt.title('price_range Distribution by n_cores')
plt.xlabel('n_cores')
plt.ylabel('price_range')
plt.xticks(rotation=45) # Rotate x-axis labels if needed
plt.show()



```
