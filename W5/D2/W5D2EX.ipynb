A LLM (Large Language Model) is an AI model trained on massive text data to understand and generate human-like language. Built on Transformer architectures (e.g., GPT, BERT, LLaMA), LLMs excel at tasks like text generation, translation, summarization, and reasoning. Their power comes from billions of parameters that capture word relationships and context

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# 2. Loading a pretrained model and tokenizer
model_name = "gpt2"  # GPT-2 is used here for demonstration; can be replaced with models like "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

print(f"\nModel '{model_name}' loaded successfully!")
print("""
GPT-2 is a causal language model, meaning it predicts the next word in a sequence.
It has been trained on a diverse dataset and can generate coherent, contextually relevant text.
""")
```

Tokenization is the process of breaking down a text into smaller components called "tokens." These tokens can be words, subwords, or even individual characters, depending on the tokenizer type

```python
import matplotlib.pyplot as plt
from transformers import AutoTokenizer

# 1. Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained("gpt2")

# 2. Define text and tokenize it
text = "Tokenization is an important step in NLP."
tokens = tokenizer.tokenize(text)  # Converts text into tokens
token_ids = tokenizer.convert_tokens_to_ids(tokens)  # Converts tokens into numerical IDs

# Display tokenization results
print(f"Original Text: {text}")
print(f"Tokens: {tokens}")
print(f"Token IDs: {token_ids}")

# 3. Visualizing the tokenization process
plt.figure(figsize=(10, 4))
plt.bar(tokens, token_ids, color="skyblue")
plt.xlabel("Tokens")
plt.ylabel("Token IDs")
plt.title("Tokenization Visualization")
plt.xticks(rotation=45)
plt.show()
```

```python
for token, token_id in zip(tokens, token_ids):
    print(f"Token: {token}, ID: {token_id}")
```

Pretraining is the first phase where a Transformer model learns general language understanding from massive amounts of text.

Fine-tuning is the second phase, where the pretrained model is adapted for a specific task. This phase is usually supervised, meaning the model is trained on labeled data.

```python
text = "This is another sentence for the variable text."
```

```python
# Define input prompt
input_text = "Once upon a time, in a distant kingdom"
input_ids = tokenizer.encode(input_text, return_tensors="pt")  # Convert to token IDs

# Generate text
output_ids = model.generate(input_ids, max_length=50, do_sample=True, temperature=0.7)
output_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)

# Print results
print(f"Input: {input_text}")
print(f"Generated Output: {output_text}")
```
