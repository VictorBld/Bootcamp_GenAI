```python
from langchain.document_loaders import HuggingFaceDatasetLoader
```

```python
dataset_name = "databricks/databricks-dolly-15k"
page_content_column = "context"
```

```python
loader = HuggingFaceDatasetLoader(dataset_name, page_content_column)
data = loader.load()
print(data[:2]) # Optional: Print the first 2 entries to verify loading
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=150)
```

```python
docs = text_splitter.split_documents(data)
print(docs[0]) # Optional: Print the first document chunk
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
modelPath = "sentence-transformers/all-MiniLM-l6-v2"
model_kwargs = {'device':'cpu'}
encode_kwargs = {'normalize_embeddings': False}
```

```python
embeddings = HuggingFaceEmbeddings(
model_name=modelPath,
model_kwargs=model_kwargs,
encode_kwargs=encode_kwargs
)
```

```python
text = "This is a test document."
query_result = embeddings.embed_query(text)
print(query_result[:3])
```

```python
from langchain.vectorstores import FAISS
db = FAISS.from_documents(docs, embeddings)
```

```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering, pipeline
from langchain import HuggingFacePipeline
```

```python
tokenizer = AutoTokenizer.from_pretrained("Intel/dynamic_tinybert")
model = AutoModelForQuestionAnswering.from_pretrained("Intel/dynamic_tinybert")
```

```python
model_name = "Intel/dynamic_tinybert"
tokenizer = AutoTokenizer.from_pretrained(model_name, padding=True, truncation=True, max_length=512)
Youtubeer = pipeline(
  "question-answering",
  model=model_name,
  tokenizer=tokenizer,
  return_tensors='pt'
)
```

```python
llm = HuggingFacePipeline(
  pipeline=Youtubeer,
  model_kwargs={"temperature": 0.7, "max_length": 512},
)
```

```python
from langchain.chains import RetrievalQA
retriever = db.as_retriever(search_kwargs={"k": 4}) # Optional: You can adjust k for number of documents retrieved
```

```python
qa = RetrievalQA.from_chain_type(llm=llm, chain_type="refine", retriever=retriever, return_source_documents=False)
```

```python
question = "What is cheesemaking?"
```

```python
result = qa.run({"query": question})
print(result) # Or print(result["result"]) if the output is a dictionary
```

```python
qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=retriever, return_source_documents=False)
question = "What is cheesemaking?"

# Updated run to pass question and context to qa chain
result = qa({"query": question})
print(result)
```

```python
qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=retriever, return_source_documents=False)
question = "What is cheesemaking?"

# Get relevant documents
docs = retriever.get_relevant_documents(question)

# Format the context for the QA model
# Assuming 'page_content' contains the text content of each document
context = "\n\n".join([doc.page_content for doc in docs])

# Updated run to pass question and context as a dictionary to qa chain
result = qa({"query": question, "context": context})
print(result)
```
