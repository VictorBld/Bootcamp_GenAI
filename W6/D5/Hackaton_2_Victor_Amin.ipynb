```python
import pdfplumber

# Remplace "nom_du_fichier.pdf" par le nom du fichier téléversé
pdf_path = "les-obligations-des-salaries_.pdf"

def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + "\n"
    return text.strip()

# Extraire et afficher les 1000 premiers caractères du texte
pdf_text = extract_text_from_pdf(pdf_path)
print(pdf_text[:1000])


```

```python
!pip install sentence-transformers faiss-cpu
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

# Charger le modèle de vecteurs (BERT optimisé pour la recherche)
model = SentenceTransformer("all-MiniLM-L6-v2")

# Convertir le texte extrait du PDF en une liste de phrases
sentences = pdf_text.split("\n")

# Encoder chaque phrase en vecteurs
embeddings = model.encode(sentences)

# Initialiser FAISS et indexer les vecteurs
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(embeddings)

print("Indexation terminée. Nombre de phrases indexées :", len(sentences))
```

```python
def search(query, top_k=3):
    """Recherche les phrases les plus pertinentes dans le document."""
    query_embedding = model.encode([query])
    distances, indices = index.search(query_embedding, top_k)

    results = [sentences[i] for i in indices[0]]
    return results

query = "Quels sont les droits des salariés en entreprise ?"
relevant_sentences = search(query)
print("Résultats pertinents :\n", "\n".join(relevant_sentences))
```

```python
from transformers import pipeline

# Charger le modèle de résumé
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

def search_and_summarize(query, max_len=300, min_len=50):
    """Recherche les passages pertinents et génère un résumé optimisé."""

    # Recherche des passages pertinents
    results = search(query)
    text_to_summarize = " ".join(results)

    # Statistiques avant résumé
    original_chars = len(text_to_summarize)
    original_words = len(text_to_summarize.split())

    # Générer le résumé
    summary = summarizer(text_to_summarize, max_length=max_len, min_length=min_len, do_sample=False)
    summarized_text = summary[0]["summary_text"]

    # Statistiques après résumé
    summarized_chars = len(summarized_text)
    summarized_words = len(summarized_text.split())

    # Calcul du gain en réduction
    char_reduction = round(100 * (1 - summarized_chars / original_chars), 2)
    word_reduction = round(100 * (1 - summarized_words / original_words), 2)

    # Affichage des statistiques
    print("**Statistiques de Résumé**")
    print(f" Texte original : {original_chars} caractères, {original_words} mots")
    print(f" Résumé        : {summarized_chars} caractères, {summarized_words} mots")
    print(f" Réduction : {char_reduction}% (caractères) | {word_reduction}% (mots)\n")

    return summarized_text

# Tester avec une requête
query = "Quels sont les droits des salariés en entreprise ?"
summary = search_and_summarize(query)
print(" Résumé Final :\n", summary)
```

```python
from rouge import Rouge
rouge = Rouge()
results = search(query)
text_to_summarize = " ".join(results)
reference = text_to_summarize
candidate = summary

scores = rouge.get_scores(candidate, reference)
print("ROUGE Scores:", scores)
```

```python
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction

smooth=SmoothingFunction().method1

bleu_1 = sentence_bleu(reference, candidate, weights=(1, 0, 0, 0), smoothing_function=smooth)  # Unigrammes
bleu_2 = sentence_bleu(reference, candidate, weights=(0.5, 0.5, 0, 0), smoothing_function=smooth)  # Bigrams

print(f"BLEU-1 Score: {bleu_1:.4f}")
print(f"BLEU-2 Score: {bleu_2:.4f}")
```

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

# Charger le modèle GPT-2 pour calculer la perplexité
perplexity_model_name = "gpt2"
perplexity_tokenizer = AutoTokenizer.from_pretrained(perplexity_model_name)
perplexity_model = AutoModelForCausalLM.from_pretrained(perplexity_model_name)

def compute_perplexity(text):
    """Calcule la perplexité d'un texte avec GPT-2"""
    encodings = perplexity_tokenizer(text, return_tensors="pt", truncation=True, max_length=512)

    with torch.no_grad():
        outputs = perplexity_model(**encodings, labels=encodings["input_ids"])
        loss = outputs.loss  # La perte est directement la NLL (negative log-likelihood)

    perplexity = torch.exp(loss).item()  # Perplexité = e^(NLL)
    return perplexity
```

```python
perplexity = compute_perplexity(summary)
print(f"Perplexité du résumé : {perplexity:.2f}")
```
