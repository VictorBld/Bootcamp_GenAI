```python
def search_and_summarize(query, max_len=300, min_len=100):
    """Recherche les passages pertinents et génère un résumé optimisé."""

    # Recherche des passages pertinents
    results = search(query)
    text_to_summarize = " ".join(results)

    # Statistiques avant résumé
    original_chars = len(text_to_summarize)
    original_words = len(text_to_summarize.split())

    # Générer le résumé
    summary = summarizer(text_to_summarize, max_length=max_len, min_length=min_len, do_sample=False)
    summarized_text = summary[0]["summary_text"]

    # Statistiques après résumé
    summarized_chars = len(summarized_text)
    summarized_words = len(summarized_text.split())

    # Calcul du gain en réduction
    char_reduction = round(100 * (1 - summarized_chars / original_chars), 2)
    word_reduction = round(100 * (1 - summarized_words / original_words), 2)

    # Affichage des statistiques
    print("**Statistiques de Résumé**")
    print(f"🔹 Texte original : {original_chars} caractères, {original_words} mots")
    print(f"🔹 Résumé        : {summarized_chars} caractères, {summarized_words} mots")
    print(f"🔹 Réduction : {char_reduction}% (caractères) | {word_reduction}% (mots)\n")

    return summarized_text

# Tester avec une requête
query = "Quels sont les caractéristiques d'un match de foot?"
summary = search_and_summarize(query)
print("📌 **Résumé Final :**\n", summary)
```

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

# Charger le modèle GPT-2 pour calculer la perplexité
perplexity_model_name = "gpt2"
perplexity_tokenizer = AutoTokenizer.from_pretrained(perplexity_model_name)
perplexity_model = AutoModelForCausalLM.from_pretrained(perplexity_model_name)

def compute_perplexity(text):
    """Calcule la perplexité d'un texte avec GPT-2"""
    encodings = perplexity_tokenizer(text, return_tensors="pt", truncation=True, max_length=512)

    with torch.no_grad():
        outputs = perplexity_model(**encodings, labels=encodings["input_ids"])
        loss = outputs.loss  # La perte est directement la NLL (negative log-likelihood)

    perplexity = torch.exp(loss).item()  # Perplexité = e^(NLL)
    return perplexity
```

```python
perplexity = compute_perplexity(summary)
print(f"Perplexité du résumé : {perplexity:.2f}")
```
